## Skin Disease Classification – Exploratory Data Analysis (EDA)

The dataset used in this project consists of images representing 23 different types of skin diseases. The images were collected from the dermatology resource DermNet, a widely used online platform for dermatological education.

The dataset contains approximately 19500 images in total. These images have been pre-split into:

- Training set: 15500 images  
- Test set: 4000 images  

Each image belongs to one of 23 dermatological categories.

This dataset enables several potential machine learning tasks, including:
- Performing exploratory data analysis (EDA) to understand visual patterns and differences between dermatological conditions.
- Training a convolutional neural network (CNN) to classify images into one of the 23 skin disease categories.
- Exploring clustering or grouping methods to identify broader disease categories or similarities among conditions.

In this project, the primary goal is to develop an image classification model capable of predicting the correct skin disease category from an input image.

```python
# import libraries
import torch
from torchvision import datasets, transforms
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# set data paths
train_dir = Path("dataset/train")
test_dir = Path("dataset/test")
```

```python
# define image transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])
```

```python
# load dataset
train_data = datasets.ImageFolder(root = train_dir, transform = transform)
test_data = datasets.ImageFolder(root = test_dir, transform = transform)
```
