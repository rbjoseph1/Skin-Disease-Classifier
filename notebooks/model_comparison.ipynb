# CNN Skin Disease Classifier — Progress Brief

## Research Question

How do data augmentation and dropout regularization affect the classification performance of a CNN across 23 skin disease categories compared to a baseline architecture?

## Project Goal
Build and compare three CNN architectures for classifying skin disease images 
across 23 categories using a custom dataset.

## What's Been Completed So Far

### 1. Exploratory Data Analysis (EDA) ✅
- Analyzed dataset structure, class distribution, and image properties
- Identified class imbalance across 23 skin disease categories

### 2. Three CNN Models ✅

All three models share the same architecture:
- 4 convolutional blocks (32 → 64 → 128 → 256 filters)
- Trained on 15,557 images, tested on 4,002 images
- Each model differs by one technique only

| Model | What Changed | Accuracy | F1 Score |
|-------|-------------|----------|----------|
| Baseline CNN | Nothing — pure CNN | 26.0% | 23.0% |
| Augmented CNN | Random flips, rotations, color jitter on training data | 20.4% | 17.5% |
| Dropout CNN | 50% dropout after first classifier layer, 30% after second | 22.4% | 18.0% |

## Key Findings
- Baseline outperformed both regularization techniques
- This is expected at low epoch counts — dropout and augmentation are 
  designed to prevent overfitting over longer training runs
- All models significantly outperform random chance (4.3% for 23 classes)
- Best performing classes: Nail Fungus, Acne & Rosacea (visually distinct)
- Worst performing classes: rare/visually similar conditions with fewer samples

## Computational Constraints
Training was conducted on a MacBook CPU. To complete training overnight:
- Image size reduced: 224×224 → 64×64
- Epochs reduced: 25 → 5
- Each epoch took ~45 seconds (vs ~4.5 minutes at full resolution)

## Questions
Should we run full training (25 epochs, 224×224) on a GPU environment like 
Google Colab to get more meaningful results? Expecting dropout and augmentation to outperform baseline with longer training.
